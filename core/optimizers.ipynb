```python
import torch 
class SGD: 
    def __init__(self, params, lr= 0.1):
        self.params = params
        self.lr = lr
    def zero_grad(self):
        # change the gradients and other urnning params to be 0 
        for p in self.params: 
            if p.grad is not None : 
                p.grad.zero_()

    def step(self):
        for p in self.params: 
            p.data -= self.lr*p.grad
```

```python
import torch 
class SGD_momentum: 
    def __init__(self, params, beta = 0.9 , lr= 0.1):
        self.params = params
        self.lr = lr
        self.beta = beta 
        self.velocity = [torch.zeros_like(p) for p in self.params] 
    
    def zero_grad(self):
        # change the gradients and other urnning params to be 0 
        for p in self.params: 
            if p.grad is not None : 
                p.grad.zero_()
    def step(self):
        for i, p in enumerate(self.params):
            self.velocity[i] = self.beta * self.velocity[i] + p.grad 
            p.data -= self.lr*self.velocity[i]

```

```python
import torch 
class Adam : 
    def __init__(self, params, beta1 = 0.9 , beta2= 0.99, lr= 0.1, eps= 1e-5):
        self.eps = eps
        self.params = params
        self.lr = lr
        self.beta1 = beta1
        self.beta2 = beta2 
        self.velocity = [torch.zeros_like(p) for p in self.params] 
        self.moment = [torch.zeros_like(p) for p in self.params] 
        self.t = 0 
    def zero_grad(self):
        # change the gradients and other urnning params to be 0 
        for p in self.params: 
            if p.grad is not None : 
                p.grad.zero_()

    def step(self):
        self.t+=1
        
         
        for i, p in enumerate(self.params):
            
            self.velocity[i] = self.beta2 * self.velocity[i] + (1-self.beta2)*p.grad**2
            self.moment[i] = self.beta1 * self.moment[i] + (1-self.beta1) *p.grad 
            moment_hat  = self.moment[i] / (1-self.beta1**self.t)
            velocity_hat = self.velocity[i] / (1-self.beta2**self.t)

            
            p.data -= self.lr* moment_hat / (torch.sqrt(velocity_hat) + self.eps)

```

```python
import torch 
class AdamW : 
    def __init__(self, params, lamda = 0.01,  beta1 = 0.9 , beta2= 0.99, lr= 0.1, eps= 1e-5):
        self.eps = eps
        self.lamda = lamda
        self.params = params
        self.lr = lr
        self.beta1 = beta1
        self.beta2 = beta2 
        self.velocity = [torch.zeros_like(p) for p in self.params] 
        self.moment = [torch.zeros_like(p) for p in self.params] 
        self.t = 0 
    def zero_grad(self):
        # change the gradients and other urnning params to be 0 
        for p in self.params: 
            if p.grad is not None : 
                p.grad.zero_()

    def step(self):
        self.t+=1
        for i, p in enumerate(self.params):
            
            self.velocity[i] = self.beta2 * self.velocity[i] + (1-self.beta2)*p.grad**2
            self.moment[i] = self.beta1 * self.moment[i] + (1-self.beta1) *p.grad 
            moment_hat  = self.moment[i] / (1-self.beta1**self.t)
            velocity_hat = self.velocity[i] / (1-self.beta2**self.t)

            p.data -= self.lr * moment_hat / (torch.sqrt(velocity_hat) + self.eps)  # Adam step

            p.data -= self.lr* self.lamda * p.data

#The whole point of AdamW is that weight decay happens in addition 
# to the Adam update, not instead of it. 
```
