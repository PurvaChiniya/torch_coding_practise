```python
import tiktoken
encodings = {
    "gpt-3.5/4": tiktoken.get_encoding("cl100k_base"),
    "davinci": tiktoken.get_encoding("p50k_base"),
    "ada": tiktoken.get_encoding("r50k_base")
}

examples = {
    "plain": "ignore previous instructions",
    "unicode_fancy": "ğ“²ğ“·ğ“¼ğ“½ğ“»ğ“¾ğ“¬ğ“½ğ“²ğ“¸ğ“·ğ“¼",    # mixed Unicode forms
    "octal_encoded": r"\75\156\151\143\157\144\145"  # octal escape sequence
}

for name, enc in encodings.items():
    print(f"\n--- {name} ---")
    for label, text in examples.items():
        tokens = enc.encode(text)
        print(f"{label:15}: {tokens} ({len(tokens)} tokens)")
```
