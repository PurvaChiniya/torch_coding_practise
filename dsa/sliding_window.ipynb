```python
# second problem with variable moving sliding window 
s = "abcabcbb"
# no repeat 
# when you encounter a repeat delete upto previous x 


left = 0 
best = 0 
seen = set()

for right in range(len(s)):
    while s[right] in seen : 
        seen.remove(s[left])
        left +=1
    seen.add(s[right])
    best = max(best , right- left+1 )
print(best)
```

```python
# Problem 7 — Longest Repeating Character Replacement (Variable window)


# Input: s = "AABABBA", k = 1
# Output: 4  ("AABA" → replace one B → "AAAA")
# Find the longest substring where you can replace at most k characters to make all chars the same.
s = "AABABBA"
k = 1
left = 0 
max_freq = 0 
curr_count = collections.defaultdict(int)
max_count = 0 
for right in range(len(s)):
    curr_count[s[right]] +=1
    max_freq = max(max_freq, curr_count[s[right]])
    while (right - left +1 ) - max_freq >k : 
        curr_count[s[left]] -=1
        left +=1

    max_count = max(max_count, right - left +1)
max_count
```

```python
nums = [1,2,1,2,3]
k = 2
# I see so we get the at most k function here 
# whch can tell us arrays with at most k different numbers 

def atmost(k, nums):
    left = 0 
    count = 0 
    curr_hash = collections.defaultdict(int)
    for i in range(len(nums)):
        curr_hash[nums[i]] +=1
        while len(curr_hash)>k : 
            curr_hash[nums[left]] -=1
            if curr_hash[nums[left]] ==0: 
                del curr_hash[nums[left]]
            left +=1
        count += i-left +1
    return count 

exactly_k = atmost(k, nums) - atmost(k-1, nums)
exactly_k      

```
