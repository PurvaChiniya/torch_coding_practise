```python
# MEX partition — split array into max parts where each part's MEX matches suffix MEX
nums = [0,1,0] # [ 1, 2, 1]
n = len(nums)
suffix_mex = [0] * n
seen_suffix = [False] * (n + 2)
current_mex = 0
for i in range(n - 1, -1, -1):
    if nums[i] <= n + 1:
        seen_suffix[nums[i]] = True
    
    # Increment current_mex to find the missing number
    while seen_suffix[current_mex]:
        current_mex += 1
        
    suffix_mex[i] = current_mex

i = 0 
res = [] 
cut_id = 1 
seen_suffix = [0] * (n + 2)

while i < n : 
    current_target = suffix_mex[i]
    current_mex = 0 
    j = i 
    while j < n : 
        if nums[j] <= n + 1:
            seen_suffix[nums[j]] = cut_id 

        while seen_suffix[current_mex] == cut_id: 
            current_mex+=1
        
        if current_mex == current_target: 
            cut_id+=1
            break 
        j+=1

    i = j+1
    res.append(current_mex)
res
```
