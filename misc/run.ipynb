```python
from typing import List, Optional

class RevenueSystem:
    def __init__(self, ):
        # TODO: Initialize RevenueSystem
        self.revenue = {}
        self.customer_id = 0 
        self.reference_graph = {}

    def add(self, revenue: int) -> int:
        # TODO: Implement add logic
        self.revenue[self.customer_id] = revenue 
        self.reference_graph[self.customer_id] = []
        self.customer_id += 1 
        print(self.reference_graph, self.revenue)
         
        return self.customer_id -1


    def addByReferral(self, revenue: int, referrerId: int) -> int:
        # TODO: Implement addByReferral logic
        if referrerId not in self.reference_graph: 
            return -1 
        self.revenue[self.customer_id] = revenue
        self.revenue[referrerId] += revenue
        self.reference_graph[referrerId] += [self.customer_id]
        self.reference_graph[self.customer_id] = []
        
        self.customer_id+=1
        print(self.reference_graph, self.revenue)
        
        return self.customer_id -1

    def getTopKCustomer(self, k: int, minRevenue: int) -> List[int]:
        # TODO: Implement getTopKCustomer logic

        sorted_revenues = sorted(self.revenue.items(), key = lambda x: x[1], reverse=True)
        # get the top k 
        res = []
        print(self.reference_graph, self.revenue)
        for x, y in sorted_revenues[:k]:
            if y >= minRevenue:
                res.append(x)
        return res

    def getRelations(self, customerId: int) -> List[List[int]]:
        # TODO: Implement getRelations logic
        # given source node of customerId, find the rest of the ndoes and level of each 
        q = [customerId]
        res = []
        visited = set()
        visited.add(customerId)

        while q :
            level = len(q)
            curr_level = []
            for _ in range(level):
                node = q.pop(0)
                
                curr_level.append(node)
                for nei in self.reference_graph[node]:
                    if nei not in visited: 
                        q.append(nei)
                        visited.add(nei)
            res.append(curr_level)
        return res[1:]


 
```

```python
from typing import List, Optional
import collections 
import heapq

class CurrencyConverter:
    def __init__(self, fromArr: List[str], toArr: List[str], rateArr: List[float]):
        # TODO: Initialize CurrencyConverter
        self.graph = collections.defaultdict(list)
        for i in range(len(fromArr)):
            self.graph[fromArr[i]] += [[toArr[i], rateArr[i]]]
            self.graph[ toArr[i]] += [[fromArr[i], 1/rateArr[i]]]
            # we dont need to store it but thats fine 
        self.n = len(self.graph)

    def getBestRate(self, From: str, To: str) -> float:
        # TODO: Implement getBestRate logic
        # keep the maximum from a from to to 
        pq = [(-1.0, From)]
        best_rates = {From:1.0}
        # always keep the max on the top pop from teh max 
        result = 1 
        while pq :
            current_value, node = heapq.heappop(pq) # min heap by default 
            curr_positive_value = -1*current_value
            if node == To:
                return curr_positive_value
            print(self.graph[node])
            if  curr_positive_value < best_rates[node]:
                continue # we alraedy have a higher valeu to get to this node 

            for nei,rate in self.graph[node]:
                if (nei not in best_rates) or (curr_positive_value*rate > best_rates[nei]) :
                    best_rates[nei] = curr_positive_value*rate
                    heapq.heappush(pq,(current_value*rate, nei))
        return -1
```
