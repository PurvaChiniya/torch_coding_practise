```python
from mcp.server.fastmcp import FastMCP

# 1. Initialize the server
mcp = FastMCP(
    name="MyMathServer",
)
```

```python
# 2. Define a tool using a simple decorator
@mcp.tool()
def add(a: int, b: int) -> int:
    """
    Adds two integers together and returns the sum.
    """
    print(f"Server was asked to add {a} and {b}")
    return a + b

# 3. Define a resource (data the AI can read)
@mcp.resource("file://my_notes.txt")
def get_notes() -> str:
    """
    Returns the content of the my_notes.txt file.
    """
    try:
        with open("my_notes.txt", "r") as f:
            return f.read()
    except FileNotFoundError:
        return "Note file not found."
```

```python
mcp.run(transport="stdio")
```
