```python
import random
tens = [10] * 16 
others = [val for val in range(2, 10) for _ in range(4)]
aces = [11] * 4
deck = tens + others + aces
print(len(deck))
def deal(deck):
    print(f"Deck has {len(deck)} cards.")
    # 2. Pop the last item
    # shuffle 
    random.shuffle(deck)
    card1 = deck.pop()
    card2 = deck.pop()
    return card1, card2 

def calculate_sum(hand):
    num_aces = hand.count(11)
    sum_ = sum(hand)
    while sum_ > 21 and num_aces > 0:
        sum_ = sum_ - 10     # Change one Ace from 11 to 1
        num_aces = num_aces - 1   # "Use up" one of our Aces
    has_ace = (num_aces > 0)
    return sum_, has_ace
def get_state(player_hand, dealer_card):
    # 1. Call your calculate_sum function
    player_sum, has_ace = calculate_sum(player_hand)
    # 2. Return the state tuple
    return (player_sum, dealer_card, has_ace)
# lets start here with s0 
ACTIONS = ["Hit", "Stand"]
q_table = {} 
epsilon = 0.9
def choose_action(state):
    # this takes the input as the (player_sum, dealer_card, has_ace) fucntion and you need to 
    if state not in q_table:
        q_table[state] = {}
        for action in ACTIONS:
            q_table[state][action] = 0.0  #
    player_sum, dealer_card, has_ace = state 
    if random.random() < epsilon:
        # explore 
        return random.choice(ACTIONS)
    else:
        
        state_q_values = q_table[state]
        # Find the action (key) that has the maximum Q-value (value)
        best_action = max(state_q_values, key=state_q_values.get)
        return best_action
def update_q_table(q_table, state, action, reward, new_state, done, alpha, gamma):
    old_value = q_table[state][action]
    if done:
        max_future_q = 0.0
    else:
        # If the new_state is one we've never seen, initialize it.
        if new_state not in q_table:
            q_table[new_state] = {}
            for act in ACTIONS:
                q_table[new_state][act] = 0.0
        max_future_q = max(q_table[new_state].values())
    new_value_target = reward + (gamma * max_future_q)
    new_q_value = old_value + alpha * (new_value_target - old_value)
    q_table[state][action] = new_q_value
def get_new_deck():
    """Creates and shuffles a new 52-card deck."""
    tens = [10] * 16 
    others = [val for val in range(2, 10) for _ in range(4)]
    aces = [11] * 4
    deck = tens + others + aces
    random.shuffle(deck)
    return deck
```
