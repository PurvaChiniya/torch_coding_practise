```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt # for making figures
%matplotlib inline
# download the names.txt file from github
words = open('names.txt', 'r').read().splitlines()
words = words[:100]
```

```python
chars = sorted(list(set(''.join(words))))
print(len(chars))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s,i in stoi.items()}
vocab_size = len(itos)
# dataset construction 
block_size = 3
# three gram context at character level 
def create_dataset(words, block_size):
    X = []
    Y = []
    for w in words: 
        # initnal context 
        context = [0]*block_size
        for char in w +".":
            # append the new one and make the next to label 
            ix = stoi[char]
            X.append(context)
            Y.append(ix)
            context = context[1:] + [ix]    
    X= torch.tensor(X)
    Y = torch.tensor(Y)
    
    return X, Y
        
block_size = 3
n = len(words)
train_words = words[:int(0.7 * n)]
val_words   = words[int(0.7 * n):int(0.85 * n)]
test_words  = words[int(0.85 * n):]

train_data, train_labels = create_dataset(train_words, block_size)
val_data, val_labels = create_dataset(val_words, block_size)
test_data, test_labels = create_dataset(test_words, block_size)

print(train_data.shape)
print(val_data.shape)
print(test_data.shape)
```
